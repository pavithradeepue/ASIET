```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
# Step 1: Load the dataset
df = pd.read_csv("Live.csv")
```

```python
# Select features for clustering
features = ['num_reactions', 'num_comments', 'num_shares', 'num_likes',
            'num_loves', 'num_wows', 'num_hahas', 'num_sads', 'num_angrys']
X = df[features].values
```

```python
# Step 2: Apply KMeans clustering
kmeans = KMeans(n_clusters=4, random_state=0)
kmeans.fit(X)
```

```python
# Step 3: Get cluster assignments and centroids
y_kmeans = kmeans.predict(X)
centers = kmeans.cluster_centers_
```

```python
# Step 4: Plot results using first two features (e.g., num_reactions and num_comments)
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.75, marker='X', label='Centroids')
plt.xlabel('num_reactions')
plt.ylabel('num_comments')
plt.title('K-Means Clustering on Live.csv Dataset')
plt.legend()
plt.show()
```
